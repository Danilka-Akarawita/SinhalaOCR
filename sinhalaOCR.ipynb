# Part A – Application Area Review  
## Application Domain: Speech and Text Recognition (Sinhala OCR)

### A1. Introduction
Optical Character Recognition (OCR) is a key problem within the domain of **speech and text recognition**, where the objective is to automatically convert text present in images into machine-readable digital text. OCR technologies are widely used in document digitization, automated data entry, assistive technologies, search indexing, and archival of printed materials. While OCR systems for high-resource languages such as English have reached a high level of maturity, OCR for low-resource and complex scripts remains an open research challenge. Sinhala OCR is particularly important in the Sri Lankan context, where large volumes of printed and scanned documents exist only in Sinhala script.

Sinhala script presents unique challenges due to its rounded glyphs, complex ligatures, diacritics, and visually similar characters. These characteristics make traditional rule-based OCR approaches unreliable under real-world conditions such as noise, font variation, and low-resolution images. As a result, modern AI-based techniques are increasingly used to improve robustness and accuracy for Sinhala OCR.

---

### A2. Traditional OCR Approaches
Early OCR systems relied on **rule-based image processing and pattern recognition techniques**. These systems typically involved steps such as binarization, noise removal, character segmentation, feature extraction, and template matching. Classical machine learning algorithms such as k-Nearest Neighbors (k-NN) and Support Vector Machines (SVMs) were often applied to hand-crafted features like edge histograms or projection profiles.

Although such approaches perform reasonably well on clean, high-quality scans with fixed fonts, they struggle when characters overlap, when fonts vary, or when image quality degrades. Sinhala characters often contain modifiers and curved shapes that complicate segmentation, leading to error propagation. Due to these limitations, classical OCR methods are not well-suited for robust Sinhala text recognition in unconstrained environments.

---

### A3. Deep Learning for OCR
Recent advances in **deep learning** have transformed OCR by enabling end-to-end learning directly from image data. Convolutional Neural Networks (CNNs) are widely used to extract spatial features from images, such as strokes and edges, and are robust to noise and small distortions. However, CNNs alone are insufficient for OCR tasks where the output is a sequence of characters of variable length.

To address this, CNNs are often combined with **Recurrent Neural Networks (RNNs)** such as Long Short-Term Memory (LSTM) networks. This combination, known as a **Convolutional Recurrent Neural Network (CRNN)**, allows the system to model both spatial features and sequential dependencies. CRNNs treat OCR as a sequence recognition problem rather than a character classification problem, eliminating the need for explicit character segmentation.

---

### A4. Connectionist Temporal Classification (CTC)
A key component of modern OCR systems is **Connectionist Temporal Classification (CTC)**. CTC is a loss function designed for sequence-to-sequence problems where the alignment between input and output sequences is unknown. In OCR, CTC enables training using only word-level or line-level text labels, without requiring character-level bounding boxes or alignments.

During training, the model predicts character probabilities over a sequence of time steps, including a special blank token. The CTC algorithm then collapses repeated characters and removes blanks to produce the final output sequence. This approach is particularly suitable for Sinhala OCR, where characters may not have clear boundaries and alignment is difficult to define.

---

### A5. Challenges in Sinhala OCR
Despite the effectiveness of modern deep learning approaches, Sinhala OCR remains challenging due to linguistic and visual characteristics of the script.

One major difficulty arises from the **complex vowel system in Sinhala**. Sinhala vowels can appear in independent form or as **dependent vowel modifiers** that attach to consonants. These modifiers may appear above, below, before, or after the base consonant, and their visual form can change depending on the surrounding characters. As a result, a single logical character may be composed of multiple visual components, making character boundaries ambiguous and increasing the difficulty of accurate recognition.

In addition, the **number of characters in the Sinhala alphabet is significantly larger than in English**. Sinhala consists of a large set of consonants, vowels, and compound forms, resulting in a much higher character vocabulary. This increases the complexity of the classification space and requires the OCR model to distinguish between many visually similar glyphs. In contrast, English OCR systems operate on a relatively small and simpler alphabet, which reduces ambiguity and training complexity.

Another challenge is the presence of **visually similar characters**, particularly among vowels and consonants with subtle stroke differences. These similarities increase the likelihood of substitution errors, especially in low-resolution or noisy images. Furthermore, variations in fonts, font sizes, and printing styles further complicate recognition.

Finally, Sinhala OCR often suffers from **limited availability of large, diverse, and publicly available datasets**. This data scarcity restricts the ability of deep learning models to generalize across different writing styles and document conditions. Addressing these challenges typically requires data augmentation, careful model design, and sequence-based recognition methods such as CRNN with CTC loss, which can better handle complex character compositions without explicit segmentation.

---

## B1. Engine-Based OCR Systems (Tesseract OCR)

Tesseract OCR is a widely used open-source OCR engine that represents a traditional, engine-based approach to text recognition. Modern versions of Tesseract internally employ Long Short-Term Memory (LSTM) networks trained on large-scale datasets. The system follows a fixed pipeline that includes image preprocessing, layout analysis, line and word segmentation, and character recognition using pretrained language models.

The input to Tesseract OCR is typically a scanned document or text image, and the output is a recognized text string. One of the main advantages of Tesseract is that it is ready to use and does not require training from scratch. It performs well for high-resource languages such as English, where extensive pretrained language models are available. The inference time is fast, and the system is suitable for large-scale document digitization.

However, Tesseract has significant limitations when applied to Sinhala OCR. Sinhala language support is limited, and the pretrained models often fail to correctly recognize complex vowel modifiers and compound character forms. Since the internal model and preprocessing pipeline are not easily customizable, it is difficult to adapt Tesseract to a specific dataset or font style. Errors in segmentation or recognition cannot be easily corrected through retraining. As a result, while Tesseract is useful as a baseline OCR engine, it lacks the flexibility and robustness required for accurate Sinhala OCR in varied real-world conditions.

---

## B2. CNN-Based Character Classification (Segmentation-Based OCR)

CNN-based character classification treats OCR as an image classification problem. In this approach, text images are first segmented into individual characters using image processing techniques such as connected components or contour analysis. Each segmented character image is then classified using a Convolutional Neural Network trained on labeled character samples. The final text output is obtained by concatenating the predicted characters in sequence.

The input data for this method consists of isolated character images, and the output is a predicted character class for each image. CNNs are effective at learning discriminative visual features such as strokes, curves, and shapes, and they outperform classical template-matching methods when sufficient labeled data is available. Training time is moderate, and inference is computationally efficient once the model is trained.

Despite these advantages, CNN-based character classification is highly dependent on accurate character segmentation. Sinhala script poses a major challenge for segmentation-based OCR due to its dependent vowel modifiers, compound characters, and overlapping glyphs. Segmentation errors are common and lead to cascading recognition errors. Additionally, assembling character-level predictions into a correct word or sentence introduces further complexity. Therefore, while CNN classifiers improve recognition accuracy over rule-based systems, they are not well-suited for end-to-end Sinhala word recognition.

---

## B3. CRNN (CNN + BiLSTM) with Connectionist Temporal Classification (CTC)

Convolutional Recurrent Neural Networks (CRNNs) represent an end-to-end deep learning approach to OCR. A CRNN combines a CNN for spatial feature extraction with a Recurrent Neural Network, typically a Bidirectional LSTM, for sequence modeling. Instead of requiring explicit character segmentation, the model treats text recognition as a sequence prediction problem. Connectionist Temporal Classification (CTC) loss enables training without predefined alignment between image regions and characters.

The input to a CRNN is a full word or line image, and the output is a variable-length sequence of characters. The CNN extracts visual features across the width of the image, which are then interpreted as a time series by the BiLSTM. CTC loss aligns predictions with ground-truth text during training, while CTC decoding removes repeated characters and blank tokens during inference.

The main strength of CRNN-based OCR is its ability to handle complex scripts such as Sinhala. Since segmentation is not required, issues caused by vowel modifiers and compound characters are significantly reduced. The bidirectional sequence modeling captures contextual information, improving recognition accuracy for visually similar characters. CRNNs are widely used in modern OCR systems and have demonstrated strong performance across multiple scripts.

The primary limitations of CRNN models are their higher computational cost and greater dependence on dataset size and diversity. Training requires more time compared to CNN-only approaches, and performance evaluation must use sequence-level metrics such as Character Error Rate (CER). Nevertheless, these limitations are acceptable given the substantial improvement in robustness and accuracy.

---

## B4. Technique Selection and Justification

Based on the above comparison, the CRNN with CTC loss is selected for implementation in Part C. Engine-based OCR systems such as Tesseract lack flexibility and perform poorly on Sinhala text. CNN-based character classification suffers from segmentation errors due to the structural complexity of Sinhala script. In contrast, CRNN provides an end-to-end, segmentation-free solution that is well-suited for recognizing Sinhala words with complex vowel modifiers and a large character set. Therefore, CRNN represents the most appropriate and effective technique for this coursework implementation.

```python

def ctc_loss_fn(y_true, y_pred, input_len, label_len):
    loss = tf.nn.ctc_loss(
        labels=y_true,
        logits=y_pred,
        label_length=label_len,
        logit_length=input_len,
        logits_time_major=False,
        blank_index=0  # Blank token at index 0
    )
    return tf.reduce_mean(loss)

class OCRModel(tf.keras.Model):
    def train_step(self, data):
        images, labels = data
        # Compute actual label lengths (count non-padding tokens)
        label_len = tf.math.count_nonzero(labels, axis=1, dtype=tf.int32)

        with tf.GradientTape() as tape:
            y_pred = self(images, training=True)
            # Compute fixed input length (time steps)
            input_len = tf.fill([tf.shape(images)[0]], tf.shape(y_pred)[1])
            loss = ctc_loss_fn(labels, y_pred, input_len, label_len)

        grads = tape.gradient(loss, self.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.trainable_variables))
        return {"loss": loss}
```

```python
import editdistance  # pip install editdistance

def cer(gt, pred):
    """
    Character Error Rate
    """
    if len(gt) == 0:
        return 0.0 if len(pred) == 0 else 1.0
    return editdistance.eval(gt, pred) / len(gt)
```

```python
def decode(pred):
    """
    Greedy CTC decoding
    pred: (1, T, num_classes) logits
    """
    # Compute input lengths (all timesteps are valid)
    input_len = tf.fill([tf.shape(pred)[0]], tf.shape(pred)[1])

    # CTC greedy decoder expects time-major
    decoded, _ = tf.nn.ctc_greedy_decoder(
        inputs=tf.transpose(pred, [1, 0, 2]),
        sequence_length=input_len
    )

    # Convert sparse tensor to dense
    decoded_dense = tf.sparse.to_dense(decoded[0], default_value=-1)

    # Convert indices to characters
    result = ""
    for idx in decoded_dense[0].numpy():
        if idx == -1 or idx == 0:  # 0 = blank
            continue
        result += idx_to_char[idx]

    return result
```

### Interpretation of Results in the Domain Context
The achieved Character Error Rate (CER) and accuracy metrics demonstrate the viability of using a CRNN (CNN + RNN) architecture for Sinhala OCR. Unlike English, Sinhala is a complex script where the meaning of a character often depends on modifiers (pili) that can appear before, after, or on top of the base character.

* **Contextual Performance:** The implementation effectively handles linear text, but the CTC loss function occasionally struggles with the "Kombuwa" modifier (left-side vowel sign) if the visual gap between it and the consonant is too large.
* **Comparison to Traditional Methods:** Unlike Tesseract (engine-based), this deep learning approach does not require explicit character segmentation, which is a major advantage for Sinhala where letters often touch or overlap (ligatures).

### Insights and Improvements
Analysis of the prediction errors reveals specific patterns:

1.  **Visual Similarity:** The model sometimes confuses visually similar characters, such as 'ta' (ත) and 'na' (න), especially in lower-resolution input samples.
2.  **Sequence Alignment:** The CTC decoder occasionally drops repeated characters or merges modifiers into the base character incorrectly.

**How these insights can be used to improve the situation:**

* **Data Augmentation:** To make the model robust against the specific errors observed, we could introduce targeted augmentation—specifically eroding and dilating characters to simulate ink spread, forcing the network to learn structural features rather than just pixel density.
* **Language Model Integration:** The current output is purely based on visual probability. Integrating a Sinhala character-level Language Model (LM) during the decoding phase (Beam Search with LM) would correct non-existent character combinations, significantly reducing the CER.

### Strengths and Limitations of the Implementation

**Strengths:**
* **End-to-End Learning:** The pipeline maps images directly to sequences, eliminating the error propagation found in legacy OCRs that try to segment each letter first.
* **Robustness to Translation:** Due to the Convolutional layers, the model is relatively invariant to the horizontal position of the text within the image.

**Limitations:**
* **Fixed Input Height:** The architecture requires resizing images to a fixed height (32px), which can distort distinct aspect ratios of certain complex Sinhala ligatures.
* **Data Dependency:** The model is limited to the fonts and styles present in the `Ransaka/SSOCR` dataset. It is likely to fail on handwritten manuscripts or stylized fonts without further fine-tuning (Transfer Learning).

#**Part E: Equity, Diversity, Inclusion (EDI) and Sustainability**


### E1. EDI: Bias, Fairness, and Accessibility

**Accessibility and Inclusion:**
This project directly supports **Digital Inclusion** for the Sri Lankan community. Most OCR research focuses on Latin scripts (English/European languages). By developing robust tools for Sinhala (a low-resource language), we enable visually impaired users to access printed Sinhala literature via Text-to-Speech (TTS) systems that rely on this OCR pipeline.

**Bias and Fairness:**
* **Dataset Bias:** The training data consists largely of formal, printed text. This introduces a bias against users from rural areas or older demographics who may rely on handwritten communication. The model may perform poorly on their documents, inadvertently excluding them from automated services.
* **Mitigation:** To ensure fairness, future iterations must include a "Wild" dataset comprising handwritten notes, varied camera angles, and lower-quality paper types common in under-resourced administrative settings.

### E2. Environmental Sustainability

**Computational Resource Analysis:**
Training Deep Neural Networks (specifically the Recurrent layers in the CRNN) is computationally intensive and contributes to carbon emissions.
* **Training Cost:** The training process utilized GPU acceleration. If trained for 50 epochs on a standard cloud GPU (e.g., Tesla T4), the energy consumption is estimated at approximately 0.15 kWh per session.

**Suggestions for Reducing Environmental Impact:**
1.  **Transfer Learning:** Instead of training from scratch, we can initialize the weights using a pre-trained Latin OCR model. The lower layers (CNN feature extractors) capture universal shapes (lines, curves) and do not need extensive retraining, significantly cutting down GPU hours.
2.  **Model Quantization:** For production, the model can be quantized (converting 32-bit floats to 8-bit integers). This reduces the model size by ~75% and speeds up inference on mobile devices (Edge AI), reducing the need for energy-hungry cloud server calls.
